# reindex , intersection do wyszukiwania list

```python
countries.dropna(inplace = True)
```

```python
countries.head(20)
```

```python
toFind = ['BB','AA','BS']
# AA nie znajduje sie w dataseries countries
```

```python
countries.loc[toFind]
# blad
```

```python
countries.reindex(toFind) 
# to pozwala wyswietlic calosc i N/A dla indexu AA, dziala tylko dla niezduplikowanych indexow
```

```python
countries.index.intersection(toFind)
# to wyswietla czesc wspolna data series oraz listy z indexami
```

```python
countries.loc[countries.index.intersection(toFind)]
# to pozwala wyswietlic czesci wspolne bez N/A, gdyby byl jakis duplikat to zwroci dane podwojnie
```
