# isin, isnull, notnull, between

```python
fuel = pd.read_csv(r'C:\Users\maxym\Desktop\cheatsheets\pandas_anaconda_numpy\course-files\fuel.csv',low_memory = False, usecols=['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)','Start Stop Technology'], index_col='Vehicle ID')
fuel.head()
```

```python
listofCars = ['Renault','Toyota','Ford']
```

```python
isinList = fuel['Make'].isin(listofCars)
```

```python
isinList.head()
```

```python
fuel[fuel['Make'].isin(listofCars)].head()
```

```python
isNoStartStopDef = fuel['Start Stop Technology'].isnull()
```

```python
fuel[isNoStartStopDef].head()
```

```python
isStartStopDefined = fuel['Start Stop Technology'].notnull()
isStartStopDefined.head()
```

```python
fuel[isStartStopDefined].head()
```

```python
mpg50_60 = fuel['Combined MPG (FT1)'].between(50,60)
mpg50_60.head()
```

```python
fuel[mpg50_60].head()
```

```python
print(f'There are {len(fuel[isinList])} cars within the list {listofCars}')
print('There are %d cars where Start/Stop technology is not determined' %
len(fuel[isNoStartStopDef]))
print('There are %d cars where Start/Stop technology is determined' %
len(fuel[isStartStopDefined]))
print('There are %d cares with MPG value between 50 and 60' %
len(fuel[mpg50_60]))
```
