# string operations

```python
fortune = pd.read_csv(r'C:\Users\maxym\Desktop\cheatsheets\pandas_anaconda_numpy\course-files\Fortune_500_2017.csv',usecols=['Rank','Title','Industry',
'Hqlocation','Sector'])
fortune.head(5)
```

```python
fortune["Sector"] = fortune["Sector"].str.upper()
fortune.head()
```

```python
fortune[['City','State']] = fortune["Hqlocation"].str.split(",",expand=True)
fortune.head()
```

```python
def BuildShortcut(row):
    industry = row["Industry"]
    result = ''
    for i in industry.split(' '):
        result += i[0]
    return result
```

```python
BuildShortcut({'Industry':'Factory Under Newspaper'})
```

```python
fortune['IndustryShort'] = fortune.apply(BuildShortcut,axis=1)
fortune.head()
```
