# stack unstack

```python
fuel = pd.read_csv(r'C:\Users\maxym\Desktop\cheatsheets\pandas_anaconda_numpy\course-files\fuel.csv',
                   usecols=['Year','Make','Model','Range (FT1)','City Range (FT1)'])

fuel.sort_index(inplace=True)
fuel.head(10)
```

```python
fuel.drop_duplicates(subset=['Year','Make','Model'],inplace=True)
fuel.dropna(how='any',inplace=True)
fuel.set_index(['Year','Make','Model'],inplace=True)
```

# pivot table

```python
df = pd.read_csv(r'C:\Users\maxym\Desktop\cheatsheets\pandas_anaconda_numpy\course-files\marathon_results_2016.csv',
                index_col='Bib',usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
df = df[(df['40K'] != '-') & (df['Half'] != '-')]
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df.head()
```

```python
df.pivot_table(index="Age",columns="M/F",values="TotalSeconds", aggfunc='mean').head
```

```python
df.pivot_table(index="Age",columns="M/F",aggfunc='count',values='City').head()
```

```python
df.pivot_table(index="Country",aggfunc='count',values='City').head()
```

```python
df.pivot_table(index="Country",aggfunc='count',values='City'
 ).sort_values('City', ascending=False).head()
```

```python
df.pivot_table(index=["Country","State","City"],aggfunc='mean',
values='TotalSeconds').head()
```

```python
df.pivot_table(index=["Country","State","City"],aggfunc='mean',values='TotalSeconds' ).sort_values('TotalSeconds', ascending=False).head()
```
