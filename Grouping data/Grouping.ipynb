# intro to grouping

```python
df = pd.read_csv(r'C:\Users\maxym\Desktop\cheatsheets\pandas_anaconda_numpy\course-files\marathon_results_2016.csv',
                index_col='Bib',usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
df = df[(df['40K'] != '-') & (df['Half'] != '-')]
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df.head()
```

```python
cities = df["City"].unique()
```

```python
groups = {}
```

```python
for x in cities:
 SubDfCities = df.where(df['City'] == x).dropna()
 groups[x] = SubDfCities
```

```python
for city in cities:
 tmp_df = df.loc[ df["City"] == city]
 groups[city] = tmp_df
```

# agg()

```python
sex_age = df.groupby(['M/F','Age'])
```

```python
sex_age.agg({'TotalSeconds':'mean',
        'HalfSeconds':'mean'}).head()
```

```python
sex_age.agg({'TotalSeconds':'sum',
        'HalfSeconds':'sum'}).head()
```

```python
sex_age.agg({'TotalSeconds': ['sum','mean'],
        'HalfSeconds': ['sum','mean']}).head()
```

```python
funcitons = ['mean','sum','count']
```

```python
sex_age.agg({'TotalSeconds': funcitons,
            'HalfSeconds': funcitons})
```

# using of groups

```python
df = df.head(10)
group_city = df.groupby(by="City")
```

```python
for city in group_city:
 print(city[0])
```

```python
for city in group_city:
 print(city[0], len(city[1]))
```

```python
for city, city_data in group_city:
 print(city, len(city_data))
```

```python
for city, city_data in group_city:
 print(city, city_data.nsmallest(1,"TotalSeconds"))
```

```python
the_best_per_city = pd.DataFrame()
for city, city_data in group_city:
 the_best_per_city = the_best_per_city.append(city_data.nsmallest(1,"TotalSeconds"))
the_best_per_city.head(5)
```
